# Setup

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from PIL import Image
import glob

sns.set_style('darkgrid')
```

```python
# read descriptive info
with open('../data/extracted/info.csv') as f:
    desc = f.readline().split(',')

# read file info
fileinfo = pd.read_csv('../data/extracted/info.csv', skiprows=1)

# print some info
for idx, i in enumerate(desc):
    print(f'{fileinfo.columns[idx]} ===== {i}')
print(f'Number of rows: {fileinfo.shape[0]}')
fileinfo.head()
```
